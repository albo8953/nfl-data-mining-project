> All about Possession times -

In this case our possession time is in HH:MM:SS format, and hence the following function definition format. Using the one we used during games_test, which was in MM:SS would break this, that is why we are not using that function definition.

```python
import re
def time_to_seconds(t):
    if not isinstance(t, str):
        return None
    # Extract only digits and colons (ignore text like "OT", "PT34M52S", etc.)
    cleaned = re.findall(r"[0-9:]+", t)
    if not cleaned:
        return None
    cleaned = cleaned[0]  # first match
    parts = cleaned.split(":")
    # Handle MM:SS
    if len(parts) == 2:
        m, s = parts
        return int(m)*60 + int(s)
    # Handle HH:MM:SS
    if len(parts) == 3:
        h, m, s = parts
        return int(h)*3600 + int(m)*60 + int(s)
    return None
```

```python
df['h_poss_time_sec'] = df['h_poss_time'].apply(time_to_seconds)
df['a_poss_time_sec'] = df['a_poss_time'].apply(time_to_seconds)
```

> Why do we need possession time at all?

Possession time is one of the most meaningful game‑level features in NFL analytics. Converting it to seconds gives us a numeric variable we can use for:

A. Visualizations -

We can’t plot "34:52" as a number.
You can plot 2092 seconds.
This enables:
- scatterplots (possession time vs points)
- regression lines
- correlations
- distributions
- comparisons across seasons

B. Modelling

Machine learning models cannot use "34:52" as a string.
They can use 2092 as a numeric feature.
Possession time is often predictive of:
- scoring
- total yards
- win probability
- offensive efficiency

C. Cleaning & consistency

The API returns possession time in multiple formats:
- MM:SS
- HH:MM:SS
- 34:52 (OT)
- PT34M52S
- "00:00:00"

If we don’t convert them, our dataset becomes inconsistent and unusable for analysis.
